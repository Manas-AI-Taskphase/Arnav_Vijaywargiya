```python
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv("train.csv")
df = df.drop(["Name"],axis=1)
df=df.replace('male', 1)
df=df.replace('female',0)

# test data set
df2 = pd.read_csv("test.csv")
df2 = df2.drop(["Name"],axis=1)
df2=df2.replace('male', 1)
df2=df2.replace('female',0)


train_data = df.to_numpy()
Y = train_data[:,0:1].reshape(train_data.shape[0],1)
X = train_data[:, 1:]

test_data = df2.to_numpy()
Y_test = test_data[:,0:1].reshape(test_data.shape[0],1)
X_test = test_data[:,1:]
```

```python
class LogisticRegression:
    def __init__(self, learning_rate=0.00001, num_iterations=1000000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def initialize_parameters(self, num_features):
        self.weights = np.zeros((num_features, 1))
        self.bias = 0

    def cost_function(self, y, y_pred):
        m = len(y)
        cost = -(1/m) * np.sum(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
        return cost

    def fit(self, X, y):
        num_features = X.shape[1]
        self.initialize_parameters(num_features)

        for i in range(self.num_iterations):
            z = np.dot(X, self.weights) + self.bias
            y_pred = self.sigmoid(z)

            dw = (1 / len(y)) * np.dot(X.T, (y_pred - y))
            db = (1 / len(y)) * np.sum(y_pred - y)

            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db


            if i % 100 == 0:
                cost = self.cost_function(y, y_pred)
                print(f"Iteration {i}, Cost: {cost}")
                print("Accuracy is: ", self.accuracy(y, self.predict(X)))

    def accuracy(self,Y,Y_pred):
        corr=0
        for i in range(0,Y.shape[0]):
            if Y[i][0] == Y_pred[i][0]:
                corr+=1
        return corr/Y.shape[0]
            

    def predict(self, X):
        z = np.dot(X, self.weights) + self.bias
        y_pred = self.sigmoid(z)
        predictions = (y_pred >= 0.5).astype(int)
        return predictions
```

```python
logistic_regression = LogisticRegression()
logistic_regression.initialize_parameters(X.shape[1])
logistic_regression.fit(X,Y)
predictions = logistic_regression.predict(X)
print(predictions)
```

```python
print(logistic_regression.predict(X_test))
```

```python
print("Accuracy is: ", logistic_regression.accuracy(Y_test, logistic_regression.predict(X_test)))
```
